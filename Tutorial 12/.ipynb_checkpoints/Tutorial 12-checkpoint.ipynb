## Bosques Aleatorios (RF)

En el tutorial pasado estudiamos cmo implementar una Mquina de Soporte Vectorial (SVM), la cul, naturalmente puede ser bastante til para problemas de clasificacin binarios, ya que consiste en encontrar una superficie que separe el espacio asociado a dos conjuntos de datos de la mejor forma posible. Se podra resolver un problema de clasificacin multiclase entrenando un SVM por cada clase. Sin embargo, existen clasificadores cuyo algoritmo es naturalmente generalizable a una clasificacin multiclase.

Un ejemplo de este tipo de algoritmos son los *Bosques Aleatorios*. Estos son un conjunto de *rboles de Decisin* entranados de forma simultnea sobre distintas particiones del conjunto de entrenamiento. Teniendo esto en cuenta, primero entendamos el fundamento bsico de un rbol de Decisin, el cual se encuentra ilustrado a continuacin:

<div>

</div>

El nodo raz representa la entrada a nuestro modelo de clasificacin, en el cual todos los datos se encuentran agrupados. En cada nodo, los datos se dividirn segn una decisin con respecto a las entradas del vector de representacin de cada dato, dando origen a dos ramas nuevas. Los datos seguirn distintas ramas hasta alcanzar los nodos terminales, los cuales tambin se conocen como *hojas*. A cada hoja se le asigna una clase (la clase ms representada entre los datos de entrenamiento que llegaron a dicha hoja). Durante inferencia, el vector de representacin de un dato sigue el rbol de decisin desde el nodo raz hasta alguna hoja, y la etiqueta asignada corresponde a la clase asociada a la hoja en la que termin.

En este punto puede estarse preguntando, cmo se obtienen las reglas de decisin que separan los datos que llegan a cada nodo? Sin ahondar en la matemtica que se encuentra por detrs, se busca hacer divisiones en el espacio de representacin que permitan obtener las distribuciones ms "puras", es decir que separe distintas clases, procurando que los puntos de cada clase permanezcan juntos. La siguiente imagen representa este concepto:

Note que para este ejemplo aunque la primer opcin de "split" logra separar las clases verdes de las amarillas, la separacin crea subconjuntos de las otras clases involucradas, lo que no es deseable. Por otro lado, la segunda opcin de "split" separa las clases azul y verde de las clases amarilla y roja. A pesar de que no se est segregando clases individuales, se estn discriminando grupos de clases, y se estn preservando los conjuntos de cada clase, posibilitando que en otro nodo haya una discriminacin ms especfica entre menos clases y preservando la integridad de cada conjunto de mejor forma. 

Una forma sencilla de *decisin* es umbralizar un componente del vector de caractersticas de cada dato, como en el ejemplo anterior. Sin embargo, se puede construir todo tipo de superficies de separacin en el espacio de representacin.

Los bosques aleatorios buscan ponderar las decisiones tomadas por mltiples rboles para evitar el *over-fitting*. Cada rbol del bosque es entrenado con un subconjunto del *set* de entrenamiento global para evitar que todos los rboles acaben aprendiendo lo mismo.

En este caso, emplearemos un conjunto de datos ya incluido en la librera `sklearn` para estudiar cmo funciona un RF. Este conjunto de datos est construido para estudiar la tarea de clasificar flores a partir de algunas caractersticas morfolgicas. A continuacin, se carga este conjunto de datos, se muestran las dimensiones de su matriz de propiedades y vector de etiquetas asociados, y se divide en un conjunto de entrenamiento y prueba:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
print("\nLas clases consideradas en este dataset son:")
print(iris.target_names)
print("\nLos features considerados en este dataset son:")
print(iris.feature_names)

X = iris.data
print("\nLas dimensiones de la matriz X son: ", X.shape)

Y = iris.target
print("El tamao del vector Y es: ", Y.shape)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=111)

print("\nEl conjunto de entrenamiento tiene " + str(X_train.shape[0]) + " datos")
print("El conjunto de prueba tiene " + str(X_test.shape[0]) + " datos.")
```

La librera `sklearn.ensemble` nos permite crear un modelo de clasificacin a partir de un Bosque Aleatorio, entrenarlo, y despus generar inferencias sobre un conjunto de prueba en pocas lneas:

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Se crea instancia de un Bosque aleatorio. El parmetro n_estimators indica la cantidad de rboles del bosque
rf_1 = RandomForestClassifier(n_estimators = 10, random_state=42)
# Se entrena el modelo en el conjunto de entrenamiento
rf_1.fit(X_train, Y_train)
# Se obtienen las etiquetas predichas por el modelo sobre el conjunto de prueba
Y_pred_1 = rf_1.predict(X_test)
```

# Redes Neuronales

En pocas palabras, las redes neuronales pueden emplearse para obtener una funcin que mapee *optimamtente* un vector de nuestro espacio de representacin de los datos a un vector de las probabilidades de que dicho dato pertenezca a cada clase. 

Para construir dicha funcin, se concatenan mltiples capas que ejecutan el siguiente proceso:
- Reciben un vector de entrada $x$ y le aplican una transformacin lineal ($Wx + b$).
- A cada elemento del vector resultante de dicha transformacin lineal se le aplica una funcin no lineal (Ej. $\tanh{(Wx+b)}$). A esta funcin tambin se le conoce como *funcin de activacin*.

Las dimensiones del vector de entrada y salida de cada capa est dado por el tamao de la capa. La "*magia*" de las redes neuronales yace en que a travs de mtodos de optimizacin como el descenso de gradiente (estudiado en el tutorial pasado) e implementando la regla de la cadena, se pueden optimizar las matrices de parmetros $W$ y $b$ de cada capa. Lo anterior, nos permite crear una funcin que optimice el mapeo de descriptores a clases sobre un conjunto de entrenamiento.

Para estudiar la implementacin de una red neuronal y su uso como clasificador partiremos del conjunto de datos ilustrado a continuacin. Dicho conjunto consiste de puntos representados por dos caractersticas, y los cules pertenecen a dos posibles clases (0 o 1 representadas por rojo y azul respectivamente).

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import sklearn.datasets
import sklearn.linear_model

def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y, cmap=plt.cm.Spectral)

def load_planar_dataset():
    np.random.seed(1)
    m = 400 # number of examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros((m,1), dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 4 # maximum ray of the flower

    for j in range(2):
        ix = range(N*j,N*(j+1))
        t = np.linspace(j*3.12,(j+1)*3.12,N) + np.random.randn(N)*0.2 # theta
        r = a*np.sin(4*t) + np.random.randn(N)*0.2 # radius
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        Y[ix] = j
        
    X = X.T
    Y = Y.T

    return X, Y
```

A continuacin, crearemos nuestra red neuronal. Para eso, implementaremos una funcin auxiliar que nos permita extraer algunas dimensiones importantes de la red, y luego otra que inicialice las matrices $W$ con nmeros aleatorios y los vectores $b$ con ceros, esto para todas las capas de la red.

```python
def layer_sizes(X, Y):
    
    n_x = X.shape[0]
    n_h = 4
    n_y = Y.shape[0] 
    
    return (n_x, n_h, n_y)
```

```python
def initialize_parameters(n_x, n_h, n_y):
    
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros((n_y, 1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```


Ahora, implementaremos una funcin que ejecute todas las operaciones internas de la red. Es decir que reciba una matriz de descriptores, y retorne el vector con la salida de la ltima funcin sigmoide. A este proceso de calcular todas las operaciones necesarias para obtener una salida de la red se le conoce como *propagacin*.

```python
# Implementacin auxiliar que calcula la funcin sigmoide punto a punto
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def forward_propagation(X, parameters):
    
    W1, b1, W2, b2 = parameters["W1"], parameters["b1"], parameters["W2"], parameters["b2"]
    
    Z1 = np.matmul(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.matmul(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    # Se almacenan los valores intermedios de la propagacin y no solamente el valor de salida de la red
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
```

Previamente habamos mencionado que podamos encontrar los parmetros $W$ y $b$ de cada capa para clasificar ptimamente el conjunto de entrenamiento mediante el *descenso de gradiente*. Para esto, al igual que en el caso de los SVM, necesitamos una funcin de costo. La funcin de costo debe ser mnima cuando las etiquetas predichas coincidan con las etiquetas, de tal forma que al minimizar dicha funcin el desempeo del clasificador sea ptimo. Una funcin que cumple con este atributo para el problema de clasificacin binaria es la siguiente:

<div>

</div>
                                        
Note que si el valor predicho ($a^{[2](i)}$) y el valor de la etiqueta ($y^{(i)}$) son iguales y cercanos a 0 o 1, todo el trmino dentro de la sumatoria es igual a 0, lo que indicara una buena clasificacin. Por otro lado, si uno de los dos valores es cercano a 0 y el otro a 1, todo el trmino dentro de la sumatoria sera un valor negativo de gran magnitud, ya que $\scriptsize{\lim_{x \to 0^+} (\log{x}) = - \infty}$, lo que indicara una mala clasificacin. Por lo anterior, al tener un signo negativo afuera de la sumatoria garantizamos que un valor pequeo de la funcin de costo indica una buena clasificacin.

A continuacin se presenta una implementacin de la funcin de costo:

```python
def compute_cost(A2, Y, parameters):
    m = Y.shape[1]
    logprobs = np.multiply(np.log(A2),Y) + np.multiply(np.log(1 - A2), (1 - Y))
    cost = - np.sum(logprobs)/m
    cost = float(np.squeeze(cost))
    return cost
```

El mtodo de descenso de gradiente consiste en actualizar los parmetros en direccin contraria al gradiente de la funcin de costo. Podemos calcular los gradientes de la funcin de costo anterior con respecto a cada elemento de los arreglos $W_1, W_2, b_1 \, y \, b_2$ de la siguiente forma:


<div>

</div>
                                        
Observe que al utilizar la notacin de la forma $ dW^{[2]} = \frac{\partial J}{\partial W^{[2]}}$, para calcular estos gradientes es til conservar los resultados parciales de la propagacin. Esa es la razn de que nuestra funcin `forward_propagation` retorne `cache`.

Al proceso de calcular los gradientes de la funcin de costo con respecto a los parmetros se le conoce como *retropropagacin*. A continuacin se muestran dos funciones que 1) calculan los gradientes y 2) actualizan los parmetros con base en ellos.

```python
def backward_propagation(parameters, cache, X, Y):
    
    m = X.shape[1]
    
    W1, W2 = parameters["W1"], parameters["W2"]
    
    A1, A2 = cache["A1"], cache["A2"]
    
    dZ2 = A2 - Y
    dW2 = np.matmul(dZ2, A1.T) / m
    db2 = np.sum(dZ2, axis = 1, keepdims = True) / m
    dZ1 = np.matmul(W2.T, dZ2) * (1 - np.power(A1, 2))
    dW1 = np.matmul(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads
```

```python
def update_parameters(parameters, grads, learning_rate = 1.2):
    
    W1, b1, W2, b2 = parameters["W1"], parameters["b1"], parameters["W2"], parameters["b2"]
    dW1, db1, dW2, db2 = grads["dW1"], grads["db1"], grads["dW2"], grads["db2"]
    
    # Se actualizan los parmetros en direccin opuesta al gradiente con una tasa de aprendizaje
    # ingresada por parmetro
    
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

Ahora, utilizamos las funciones anteriores para entrenar nuestro modelo. Para esto, consideraremos un total de 10,000 pocas. En cada poca, se pasa todo el conjunto de entrenamiento a travs de la red, se calcula el gradiente asociado a la funcin de costo con respecto a todo este conjunto, y se actualizan los parmetros:

```python
def nn_model(X, Y, n_h, num_iterations = 10000, print_cost=False):
    # Semilla para garantizar replicabilidad
    np.random.seed(3)
    
    # Se obtienen las dimensiones de inters
    n_x = layer_sizes(X, Y)[0]
    n_y = layer_sizes(X, Y)[2]
    
    # Se inicializan los parmetros de la red con ceros y nmeros aleatorios
    parameters = initialize_parameters(n_x, n_h, n_y)
    
    # Para cada poca:
    for i in range(0, num_iterations):
        
        # 1. Propagacin
        A2, cache = forward_propagation(X, parameters)
        # 2. Clculo de la funcin de costo
        cost = compute_cost(A2, Y, parameters)
        # 3. Retropropagacin 
        grads = backward_propagation(parameters, cache, X, Y)
        # 4. Actualizacin de parmetros
        parameters = update_parameters(parameters, grads, learning_rate = 1.2)
        # Se imprime el costo despus de 1000 pocas       
        if print_cost and i % 1000 == 0:
            print ("Costo al terminar la iteracin %i: %f" %(i, cost))
    # Se retornan los parmetros del modelo entrenado
    return parameters

parameters = nn_model(X, Y, n_h = 4, num_iterations = 10000, print_cost=True)
```

Para hacer una inferencia se toman los parmetros de la red entrenada y los descriptores de los datos a inferir. Con ellos se hace una sola propagacin, y luego se aproxima el valor de salida a 0 o 1. A continuacin, se presenta una implementacin de una funcin que hace esta inferencia:

```python
def predict(parameters, X):

    A2, cache = forward_propagation(X, parameters)
    predictions = (A2 > 0.5).astype(int)
    
    return predictions
```

```python
plt.figure(figsize=(15, 8))

plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
plt.title("Borde de Decision para capa oculta de tamao " + str(4))
plt.show()
```

Note que implementar una red neuronal de ceros tiene varias complicaciones. Una de las ms importantes es obtener manualmente las expresiones de los gradientes para distintas profundidades y funciones de activacin de la red. Adems, hay que ser cuidadoso con el manejo de las dimensiones de entrada y salida de cada capa. Existen libreras que permiten implementar redes neuronales sin tener que desarrollarlas desde cero para sobrellevar estos problemas. Una de ellas es `PyTorch`, la cual fue desarrollada por Facebook (ahora Meta). 

A continuacin, se presenta un fragmento de cdigo que permite implementar la misma red neuronal anterior mediante esta librera:

```python
import sys
!{sys.executable} -m pip install torch
# Se importan los mdulos y funciones necesarias
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD

# Pytorch maneja su propio formato de arreglo, el cual se llama Tensor.
# Para poder pasar nuestros datos a travs de la red transformamos los arreglos de Numpy
# en Tensores de PyTorch

X_tensor = torch.from_numpy(X.T.astype(np.float32))
Y_tensor = torch.from_numpy(Y.T.astype(np.float32))

# Definimos nuestra red neuronal como una clase

class NeuralNetwork(nn.Module):
    def __init__(self):
        
        # Creamos cada capa como instancias de Linear
        super().__init__()
        self.fc1 = nn.Linear(2, 4, bias=True)
        self.fc2 = nn.Linear(4, 1, bias=True)
        
        # Definimos los pasos a seguir en la propagacin mediante el mtodo forward
    def forward(self,x):
        x = self.fc1(x)
        x = F.tanh(x)
        x = self.fc2(x)
        x = F.sigmoid(x)
        return x

# Creamos una instancia de nuestra clase de red neuronal
net = NeuralNetwork()

# Creamos una instancia de nuestra funcin de costo. Consultar la documentacin de torch.nn.BCELoss para
# corroborar que coincide con la frmula presentada anteriormente

criterion = nn.BCELoss()

# Se define el nmero de pocas
epochs = 10000

# Se crea una instancia de un optimizador que utilizar el mtodo de descenso de gradiente estocstico (SGD)
optm = SGD(net.parameters(), lr = 1.2)

# Creamos una funcin de entrenamiento que recibe como parmetros nuestro conjunto de entrenamiento, sus etiquetas
# el optimizador y la funcin de costo

def train(model, x, y, optimizer, criterion):
    # Se reinician los gradientes del modelo
    model.zero_grad()
    # Se hace la propagacin
    output = model(x)
    # Se calcula el costo a partir de el resultado de la propagacin y las etiquetas
    loss = criterion(output,y)
    # Se hace la retropropagacin
    loss.backward()
    # Se actualizan los valores de los parmetros del modelo segn el optimizador.
    optimizer.step()

    return loss, output

# Entrenamos el modelo por el nmero de pocas elegidas y se retorna el costo cada 1000 pocas
for i in range(epochs):
    correct = 0
    loss, predictions = train(net, X_tensor, Y_tensor, optm, criterion)
    if i % 1000 == 0:
            print ("Costo al terminar la iteracin %i: %f" %(i, loss))
            
```

Para obtener la salida de la propagacin se pasa el Tensor a travs de la red de la siguiente forma: `net(X)`. Sin embargo, para transformar los Tensores en arreglos de Numpy, dado el caso de que se quieran hacer clculos que no involucren a la librera `PyTorch`, se deben seguir unos pasos adicionales. Esto es til a la hora de hacer inferencias:

```python
# Se obtiene el Tensor de probabilidades predichas
output_tensor = net(X_tensor)

# detach() permite extraer valores de Tensores involucrados en el clculo de gradientes y numpy() los
# transforma en arreglos de numpy
output = output_tensor.detach().numpy()

# Se aproximan las probabilidades para obtener las etiquetas predichas
pred_labels = (output > 0.5).astype(int)

# Se calcula la exactituda del clasificador
correct = np.sum(pred_labels == Y.T)
total = len(Y.T)
print("La exactitud del clasificador es: ", str(correct) + " / " + str(total) + " = " + str(correct / total))
```
