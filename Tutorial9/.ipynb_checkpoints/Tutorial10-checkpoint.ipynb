# Kmeans y Clusterización
En este tutorial entenderemos el uso de K means para el analisis de imagenes en tareas de segmentación y clusterización. Esta tecnica de machine learning tiene multiples usos en el analisis de datos, en nuestro caso, los datos pueden ser pixeles (en el caso de la segmentación) o toda la imagen (en casos de clasificación o detección). Nuestro trabajo es acomodar los datos segun nos sea más conveniente. 

Como bien sabemos, KMeans es un metodo de clusterización, es decir, dado unos datos de entrenamiento y una cantidad de clusters determinada, y dado un nuevo dato, el modelo calcula un diagrama de voronoi y le asigna un cluster. 

A continuación, generaremos 200 datos de 2 dimensiones y distribuidos en 16 clusters. Veremos como funciona KMeans para clusterizar estos datos.

```python
from sklearn import cluster, datasets
from matplotlib import pyplot as plt

data=datasets.make_blobs(200,2,centers=16,random_state=23)[0] #200 datos de 2 dimensiones en 16 clusters
for i in data:
    plt.plot(i[0],i[1],'bo',)
```

```python
data
```

```python
from sklearn.cluster import KMeans
model=KMeans(2,random_state=23) #modelo de KMeans de 2 clusters
model.fit(data) #entrenar el modelo con los datos generados
etiquetas=model.predict(data) #predicciones de cluster de cada dato de entrenamiento
```

```python
colors=['bo','ro'] #colores de cada etiqueta

for i in range(len(etiquetas)):
    x,y=data[i] #coordenadas de el dato
    color_etiqueta=colors[etiquetas[i]] #color de la etiqueta de cada cluster
    plt.plot(x,y,color_etiqueta)
plt.plot(model.cluster_centers_[0][0],model.cluster_centers_[0][1],'go')
plt.plot(model.cluster_centers_[1][0],model.cluster_centers_[1][1],'go')
```

Podemos aumentar la cantidad de clusters.

```python
from sklearn.cluster import KMeans
model2=KMeans(4,random_state=23) #modelo de KMeans de 2 clusters
model2.fit(data) #entrenar el modelo con los datos generados
etiquetas=model2.predict(data) #predicciones de cluster de cada dato de entrenamiento
```

```python
colors=['bo','ro','go','yo'] #colores de cada etiqueta

for i in range(len(etiquetas)):
    x,y=data[i] #coordenadas de el dato
    color_etiqueta=colors[etiquetas[i]] #color de la etiqueta de cada cluster
    plt.plot(x,y,color_etiqueta)
```

Observe que algunos clusters más pequeños son resultado de la partición de clusters más grandes, como el cluster amarillo y el rojo. Sin embargo, esto no sucede para todos los clusters, esto no es una clusterización jerarquica. La clusterización jerarquica es una clasificación donde los clusters mayores son resultado de la combinación de clusters más pequeños.

```python
def cluster_jerarquico(data,Kmeans1,Kmeans2,Kmeans3):
    'data (ndarray): Datos de entrenamiento a predecir'
    'Kmeans1 (KMeans): Modelo Kmeans entrenado con todos los datos'
    'Kmeans2 (KMeans): Modelo Kmeans entrenado con los datos de etiqueta 0 de Kmeans1'
    'Kmeans3 (KMeans): Modelo Kmeans entrenado con los datos de etiqueta 1 de Kmeans1'
    # YOUR CODE HERE
    etiquetas_total = np.zeros(len(data), dtype=int)
    
    for i, dato in enumerate(data):
        # se usa el primer modelo
        etiqueta = Kmeans1.predict([dato])[0]
        # se usa el segundo modelo
        if etiqueta == 0:
            etiqueta2 = Kmeans2.predict([dato])[0]
        # se usa el tercer modelo
        else:
            etiqueta2 = Kmeans3.predict([dato])[0]
            etiqueta2 += 2
        # se añade la etiqueta la lista final
        etiquetas_total[i] = etiqueta2
    
    return etiquetas_total
```

```python
predicted0=Kmeans1.predict(data)
predict_aux=np.array(cluster_jerarquico(data,Kmeans1,Kmeans2,Kmeans3))
orden_0_1=data[np.where(predict_aux<2)[0]]
orden_2_3=data[np.where(predict_aux>=2)[0]]

assert len(orden_0_1)!=0,'No hay datos con etiqueta 0'
assert len(orden_2_3)!=0,'No hay datos con etiqueta 1'
assert len(np.unique(predict_aux))==4, 'Debe tener 4 etiquetas en su clasificación'
assert np.max(predict_aux)==3, 'La maxima etiqueta debe ser 3'
assert np.min(predict_aux)==0, 'La minima etiqueta debe ser 0'

assert_list_0_1=[]
assert_list_2_3=[]
for k,i in enumerate(data):
    if predicted0[k]==0 and i in orden_0_1:
        assert_list_0_1.append(1)
    elif predicted0[k]==0 and i not in orden_0_1:
        assert_list_0_1.append(0)
    if predicted0[k]==1 and i in orden_2_3:
        assert_list_2_3.append(1)
    elif predicted0[k]==1 and i not in orden_2_3:
        assert_list_2_3.append(0)
assert np.mean(assert_list_0_1)==1, 'Un dato de la clase 1 en el Kmeans1 tiene una etiqueta de 0 o 1, debería tener una etiqueta de 2 o 3'
assert np.mean(assert_list_2_3)==1, 'Un dato de la clase 0 en el Kmeans1 tiene una etiqueta de 2 o 3, debería tener una etiqueta de 0 o 1'
assert len(assert_list_0_1)+len(assert_list_2_3)==200, 'Algun dato no fue clasificado'

fig,ax=plt.subplots(1,3)

colors=['bo','ro','go','yo'] #colores de cada etiqueta

for i in range(len(etiquetas)):
    x,y=data[i] #coordenadas de el dato
    color_etiqueta=colors[predict_aux[i]] #color de la etiqueta de cada cluster
    ax[1].plot(x,y,color_etiqueta)

ax[0].set_title('2 Clusters')
colors=['bo','ro'] #colores de cada etiqueta

for i in range(len(predicted0)):
    x,y=data[i] #coordenadas de el dato
    color_etiqueta=colors[predicted0[i]] #color de la etiqueta de cada cluster
    ax[0].plot(x,y,color_etiqueta)
ax[1].set_title('4 Clusters jerarquicos')

model2=KMeans(4,random_state=23) #modelo de KMeans de 2 clusters
model2.fit(data) #entrenar el modelo con los datos generados
etiquetas=model2.predict(data) #predicciones de cluster de cada dato de entrenamiento

colors=['bo','ro','go','yo'] #colores de cada etiqueta

for i in range(len(etiquetas)):
    x,y=data[i] #coordenadas de el dato
    color_etiqueta=colors[etiquetas[i]] #color de la etiqueta de cada cluster
    ax[2].plot(x,y,color_etiqueta)
ax[2].set_title('4 Clusters')
plt.show()
```

## Segmentación por color
Como bien sabemos, las imagenes son matrices de pixeles. Si la imagen es en escala de grises, nuestros pixeles son datos de una unica dimensión (con valores de 0-1 o 0-255). Si nuestra imagen es a color, nuestros pixeles son datos de 3 dimensiones, ya que tendremos un valor para cada canal de color. 

Los datos de entrenamiento serán los siguientes:

```python
from skimage.io import imread
from skimage.transform import resize
pixel_data=[] #datos de pixeles
Im=imread('loros.jpg')
Im=resize(Im,(256,256))
x,y,z=Im.shape
for i in range(x):
    for j in range(y):
        pixel_data.append(Im[i,j,:])
        
pixel_model=KMeans(10,random_state=23).fit(pixel_data)
```

### Ejercicio 2: 
Cree una función que reciba por parametro el modelo Kmeans pixel_model y la imagen Im. Esta función debe recorrer pixel por pixel prediciendo su cluster. Luego, le asignará a el pixel la intensidad del **centroide** del cluster al que pertenece.

**Nota:** Puede usar model.cluster_centers_ para obtener la coordenada del centroide. Donde model es el modelo KMeans entrenado.

```python
from tqdm import tqdm

def Segmentation_im1(pixel_model,Im): # muy ineficiente
    'pixel_model (KMeans): Modelo KMeans entrenado con pixeles'
    'Im (ndarray): Imagen a segmentar'
    # YOUR CODE HERE
    x, y, _ = Im.shape
    new_im = np.zeros_like(Im)
    
    for i in range(x):
        for j in range(y):
            pixel = Im[i,j,:]
            cluster = pixel_model.predict([pixel])[0]
            centroid = pixel_model.cluster_centers_[cluster]
            new_im[i,j,:] = centroid

    return new_im

def Segmentation_im(pixel_model,Im):
    'pixel_model (KMeans): Modelo KMeans entrenado con pixeles'
    'Im (ndarray): Imagen a segmentar'

    # se obtiene una lista con los pixeles
    pixeles = Im.reshape(-1, Im.shape[2])

    new_im = np.zeros_like(pixeles)
    
    # se predicen los pixels de cada cluster
    clusters = pixel_model.predict(pixeles)
    # se busca el centroide de cada pixel
    for i in range(len(pixeles)):
        centroide = pixel_model.cluster_centers_[clusters[i]]
        new_im[i] = centroide
        
    # se reconstruye la imagen con los pixeles calculados
    new_im = new_im.reshape(Im.shape)
    return new_im
```

```python
new_im=Segmentation_im(pixel_model,Im)
```

```python
asser_list=[]
assert len(new_im.shape)==3, 'La imagen debería tener filas, columnas y canales'
x,y,z=new_im.shape
assert x==256 and y==256, 'Las dimensiones de la imagen no debería  cambiar'

for i in pixel_model.cluster_centers_:
    if i in new_im:
        asser_list.append(1)  
        
assert np.mean(asser_list)==1, 'Algun centroide no esta presente en la imagen'
assert (pixel_model.cluster_centers_[pixel_model.predict([Im[24,24,:]])]==new_im[24,24,:]).all(), 'El pixel 24,24 no tiene la intensidad que debe'

fig,ax=plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(Im)
ax[0].set_title('Imagen original')
ax[0].set
ax[1].imshow(new_im)
ax[1].set_title('Imagen segmentada')

for axs in ax:
    axs.set_axis_off()
```

Ya hemos segmentado nuestra imagen usando unicamente el color. Esto tiene sentido ya elementos con un mismo color pueden perteneder a un mismo elemento. Sin embargo, tambien tiene sentido que elementos cercanos unos a otros pertenecen a un mismo elemento.

```python
pixel_ij_data=[] #datos de pixeles y coordenadas
pixel_ij_model='' #modelo KMeans entrenado con pixel_ij_data
# YOUR CODE HERE

for i in range(x):
    for j in range(y):
        pixel_ij_data.append(np.concatenate([Im[i,j,:],np.array([i,j])]))
        
pixel_ij_model=KMeans(500,random_state=23).fit(pixel_ij_data)
```

```python
def Segmentation_ij_im(pixel_model,Im):
    'pixel_model (KMeans): Modelo KMeans entrenado con pixeles'
    'Im (ndarray): Imagen a segmentar'
    # YOUR CODE HERE
    
    # se obtiene una lista con los pixeles (con su respectiva posicion)
    pixeles = []
    x, y, _ = Im.shape
    for i in range(x):
        for j in range(y):
            pixeles.append(np.concatenate([Im[i,j,:],np.array([i,j])]))
    
    new_im = np.zeros((Im.shape[0]*Im.shape[1], 3))
    
    # se predicen los pixels de cada cluster
    clusters = pixel_model.predict(pixeles)
    # se busca el centroide de cada pixel, tomando solo las tres primeras coordenadas
    for i in range(len(pixeles)):
        centroide = pixel_model.cluster_centers_[clusters[i]]
        new_im[i] = centroide[:3]
        
    # se reconstruye la imagen con los pixeles calculados
    new_im = new_im.reshape(Im.shape)
    return new_im
```

Observe como la imagen resultado parecen segmentos de superpixeles que comparten una vecindad y un color determinado. Adicionalmente, observe que los superpixles tiene forma hexagonal, esto se debe a que KMeans crea clusters circulares y, al usar las coordenadas de filas,columnas en los datos de entrenamiento, el resultado final en las imagenes esto se ve representado como hexagonos.
